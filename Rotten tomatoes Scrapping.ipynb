```python
import requests
from bs4 import BeautifulSoup
```

```python
 base_site = "https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/"
```

```python
response = requests.get(base_site)
response
```

```python
html = response.content
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
with open ('Rotten_tomatoes_page.html', 'wb') as f:
    f.write(soup.prettify('utf-8'))
```

```python
divs = soup.find_all('div', {'class':'col-sm-18 col-full-xs countdown-item-content'})
#divs
```

```python
heading = [div.find('h2') for div in divs]
```

```python
heading[0]
```

```python
movie_name = [name.find('a').string for name in heading]
movie_name
```

```python
years = [name.find('span', {"class":"subtle start-year"}).string for name in heading]
```

```python
years
```

```python
clean_years = [yr.strip('()') for yr in years]
clean_years
```

```python
clean_years = [int(yr) for yr in clean_years]
clean_years
```

```python
scores = [head.find('span', {"class":"tMeterScore"}).string for head in heading]
scores
```

```python
scores = [sc.strip('%') for sc in scores]
scores
```

```python
scores = [int(sc) for sc in scores]
scores
```

```python
consensus = [div.find('div', {'class':'info critics-consensus'}) for div in divs]
#consensus
```

```python
consensus[0].text
```

```python
# first way to remove the starting Critics Consensus: from this list
```

```python
common_text = 'Critics Consensus: '
```

```python
common_len = len(common_text)
```

```python
consensus_text = [con.text[common_len:] if con.text.startswith(common_text) else con.text for con in consensus]
consensus_text
```

```python
consensus[0]
```

```python
consensus[0].contents[1].strip()
```

```python
consensus_text = [con.contents[1].strip() for con in consensus]
consensus_text
```
